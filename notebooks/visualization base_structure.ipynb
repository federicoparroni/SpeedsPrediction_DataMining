```python
# select a row
KEY = 0
```

```python
speed_row_filtered = speed_df[speed_df['KEY']==KEY]
```

```python
km_row = speed_row_filtered.KM.unique()
```

```python
def cartesian_product(*arrays):
    la = len(arrays)
    dtype = np.result_type(*arrays)
    arr = np.empty([len(a) for a in arrays] + [la], dtype=dtype)
    for i, a in enumerate(np.ix_(*arrays)):
        arr[...,i] = a
    return arr.reshape(-1, la)
```

```python
# create all the possible couples (KM, DATETIME_UTC)
cart_prod = pd.DataFrame(cartesian_product(datetimes, km_row), columns=['DATETIME_UTC', 'KM']).sort_values('KM')
```

```python
cart_prod.DATETIME_UTC = cart_prod.DATETIME_UTC.astype(str)
cart_prod.KM = cart_prod.KM.astype(int)
```

```python
cart_prod.dtypes
```

```python
pd.merge(cart_prod, speed_row_filtered ,how='left').groupby(['KM']).count()
```
