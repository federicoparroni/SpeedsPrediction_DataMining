```python
import pandas as pd
from tqdm import tqdm_notebook as tqdm
import os
import numpy as np

os.chdir('../..')

from src.utils import *
```

```python
speeds_df = pd.read_csv(resources_path('dataset', 'originals', 'speeds_train.csv.gz'))
events_df = pd.read_csv(resources_path('dataset', 'originals', 'events_train.csv.gz'))
sensors_df = pd.read_csv(resources_path('dataset', 'originals', 'sensors.csv.gz'))
```

```python
start_km = np.minimum(events_df['KM_START'].values, events_df['KM_END'].values)
end_km = np.maximum(events_df['KM_START'].values, events_df['KM_END'].values)
events_df['KM_START'] = start_km
events_df['KM_END'] = end_km
```

```python
df = events_df
print(f'Events: {len(events_df)}')
df = pd.merge(df, sensors_df, how='outer')
print(f'Merged: {len(df)}')
```

```python
df.sort_values([KEY, EVENT_TYPE]).groupby([KEY, EVENT_TYPE]).count().head(20)
```

```python
df['DISTANCE'] = df.KM - (df.KM_START + (df.KM_END - df.KM_START)/2)
reduced_df = df[[KEY, KM, EVENT_TYPE, EVENT_DETAIL, START_DATETIME, END_DATETIME, KM_END, KM_START, ROAD_TYPE, LANES, DISTANCE]]
reduced_df.sort_values([KEY, KM_START, EVENT_TYPE, START_DATETIME]).head(20)
```

```python
reduced_df.shape
#reduced_df.to_csv(resources_path('dataset', 'preprocessed', 'events_sensors.csv.gz'))
```

```python
min_distances_df = reduced_df[[KEY, KM, EVENT_TYPE, EVENT_DETAIL, START_DATETIME, END_DATETIME, KM_END, KM_START, ROAD_TYPE, LANES, DISTANCE]]
threshold = 2

# TODO: mettere una threshold in km e prendere gli elementi sotto tale threshold
min_distances_df = min_distances_df[np.abs(min_distances_df.DISTANCE) <= threshold]
min_distances_df = min_distances_df.drop_duplicates(subset=[EVENT_TYPE, EVENT_DETAIL, KM, KEY])
print(min_distances_df.shape)
min_distances_df.sort_values([KEY, EVENT_DETAIL, KM]).head(10)
# min_distances_df.to_csv(resources_path('dataset', 'preprocessed', 'min_distance_event.csv.gz'))
```

```python
speeds_df.head(10)
```

```python
dist_df = min_distances_df.merge(speeds_df, left_on=[KEY, KM], right_on=[KEY, KM])
dist_df = dist_df[(dist_df.START_DATETIME_UTC <= dist_df.DATETIME_UTC) &
                  (dist_df.END_DATETIME_UTC >= dist_df.DATETIME_UTC)]
dist_df = dist_df.merge(dist_df.groupby([EVENT_TYPE, EVENT_DETAIL])[[SPEED_AVG, N_CARS]].agg(['mean', 'std', 'count']),
                        left_on=[EVENT_TYPE, EVENT_DETAIL], right_on=[EVENT_TYPE, EVENT_DETAIL])
dist_df.head(10)
```

```python
print(dist_df.shape)
dist_df['MEAN_SPEED_AVG'] = dist_df[('SPEED_AVG', 'mean')]
dist_df['STD_SPEED_AVG'] = dist_df[('SPEED_AVG', 'std')]
dist_df['MEAN_N_VEHICLES'] = dist_df[('N_VEHICLES', 'mean')]
dist_df['STD_N_VEHICLES'] = dist_df[('N_VEHICLES', 'std')]
dist_df['COUNT'] = dist_df[('SPEED_AVG', 'count')]
event_stats_df = dist_df.drop_duplicates(subset=[EVENT_TYPE, EVENT_DETAIL])[[EVENT_TYPE, EVENT_DETAIL, 'MEAN_SPEED_AVG', 'STD_SPEED_AVG', 'MEAN_N_VEHICLES', 'STD_N_VEHICLES', 'COUNT']]
print(event_stats_df.shape)
event_stats_df = event_stats_df.sort_values(EVENT_DETAIL).fillna(0)
event_stats_df.sort_values(['EVENT_TYPE', 'EVENT_DETAIL', 'STD_SPEED_AVG'])
event_stats_df.to_csv(resources_path('dataset', 'preprocessed', 'evets_stats.csv.gz'))
```
